# Subset and plot data with 1D lon,lat coordinates using Xarray and Cartopy

With one dimensional coordinate values, we can slice these using coordinate values.  This demonstrates the power of Xarray's Pandas-like slicing.  With projected coordinates, there are 2D lon/lat arrays, so one can't slice on lon/lat.  It works here because the weather model is on a regular grid in lon/lat space. 

```python
%matplotlib inline
import xarray
import matplotlib.pyplot as plt
import numpy as np
```

```python
url = 'http://www.esrl.noaa.gov/psd/thredds/dodsC/Datasets/ghcncams/air.mon.mean.nc'
```

```python
ds = xarray.open_dataset(url)
```

```python
var='air'
dvar = ds[var]
dvar
```

```python
# slice a longitude range, a latitute range and a specific time value
lat_bnds, lon_bnds = [50, 18], [-130+360., -64+360.]
ds_cut = ds.sel(lat=slice(*lat_bnds), lon=slice(*lon_bnds), time='2015-03-01')
ds_cut
```

```python
ds_cut.time
```

```python
ds_cut.time.data.strftime("%B %Y")
```

```python
import pandas as pd 
ts = pd.to_datetime(str(ds_cut.time.data)) 
time_string = ts.strftime('%B %Y')
time_string
```

## Plot with Cartopy

```python
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

```python
crs1 = ccrs.AlbersEqualArea(central_longitude=-100)
crs2 = ccrs.PlateCarree()
```

```python
# Create a feature for States/Admin 1 regions at 1:50m from Natural Earth
states_provinces = cfeature.NaturalEarthFeature(
    category='cultural',
    name='admin_1_states_provinces_lines',
    scale='50m',
    facecolor='none')
```

```python
fig = plt.figure(figsize=(12,8))
ax = plt.axes(projection=crs2)
mesh = ax.pcolormesh(ds_cut.lon,ds_cut.lat,ds_cut.air-273.15, transform=crs2,zorder=0, vmin=-10, vmax=30)
fig.colorbar(mesh)
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS)
ax.add_feature(states_provinces, edgecolor='gray')
gl = ax.gridlines(draw_labels=True)
gl.xlabels_top = False
gl.ylabels_right = False
plt.title('{}: {}'.format(dvar.long_name, time_string));
```

```python
fig = plt.figure(figsize=(12,8))
ax = plt.axes(projection=crs1)
mesh = ax.pcolormesh(ds_cut.lon,ds_cut.lat,ds_cut.air-273.15, transform=crs2,zorder=0, vmin=-10, vmax=30)
fig.colorbar(mesh)
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS)
ax.add_feature(states_provinces, edgecolor='gray')
#gl = ax.gridlines(draw_labels=True)
gl.xlabels_top = False
gl.ylabels_right = False
plt.title('{}: {}'.format(dvar.long_name, time_string));
```
